##### Write a function rotate(ar[], d, n) that rotates arr[] of size n by d elements.  
###### https://www.geeksforgeeks.org/array-rotation/

```python
def rotate(ar,d,n):
    i=1;
    while(i<=d):
        aele = ar.pop(0);
        ar.append(aele);
        i=i+1;
    return ar;

arr = [1,2,3,4,5];
n = len(arr);
d = 6;
newar = rotate(arr,d,n);
print(newar);
```

```python
def leftRotate(arr, d, n):
    d = d % n
    g_c_d = gcd(d, n)
    for i in range(g_c_d):
         
        # move i-th values of blocks
        temp = arr[i]
        j = i
        while 1:
            k = j + d
            if k >= n:
                k = k - n
            if k == i:
                break
            arr[j] = arr[k]
            j = k
        arr[j] = temp
 
# UTILITY FUNCTIONS
# function to print an array
def printArray(arr, size):
    for i in range(size):
        print ("% d" % arr[i], end =" ")
 
# Fuction to get gcd of a and b
def gcd(a, b):
    if b == 0:
        return a;
    else:
        return gcd(b, a % b)
 
# Driver program to test above functions
arr = [1, 2, 3, 4, 5, 6, 7]
n = len(arr)
d = 2
leftRotate(arr, d, n)
printArray(arr, n)
```

# Linked List Data Structure 
###### https://www.geeksforgeeks.org/linked-list-set-1-introduction/

```python
class Node:
    def __init__(self, data):
        self.data=data;
        self.next=None;
        
class LinkedList:
    def __init__(self):
        self.head=None;
        
    def printList(self):
        temp = self.head;
        while(temp):
            print(temp.data);
            temp = temp.next;

if __name__=='__main__':
    
    llist = LinkedList();
    
    llist.head = Node(1);
    second = Node(2);
    third = Node(3);
    
    llist.head.next = second;
    second.next = third;
    
    llist.printList()
    
     
```
